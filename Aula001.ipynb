```python
# O símbolo = significa "guarda o valor"
# Não é uma equação de matemática!

nome   = "Rubens"   # caixa "nome" guarda o texto Rubens
idade  = 23        # caixa "idade" guarda o número 23
altura = 1.66      # caixa "altura" guarda 1.66

# Para ver o conteúdo de uma variável, use print()
print(nome)    # exibe: Rubens
print(idade)   # exibe: 23

# Você pode atualizar o valor de uma variável
idade = 23     # agora a caixa "idade" guarda 23
print(idade)   # exibe: 23


# int - números inteiros (sem vírgula)
qtd_alunos = 32
ano        = 2026

# float - decimais (usa ponto, não vírgula!)
peso = 72.5
altura = 1.75

# bool - apenas dois valores possíveis
aprovado = True
reprovado = False

# Descobrir o tipo de uma variável
print(type(qtd_alunos))  # <class "int">
print(type(nome))        # <class "str">

# print() exibe qualquer valor na tela
print("Olá, mundo!")
print(42)
print(nome)    # exibe o conteúdo da variável

# f-string: texto com variáveis embutidas
nome = "Rubens"
idade = 23
print(f"Meu nome é {nome} e tenho {idade} anos.")
# meu nome é rubens e tenho 23 anos.

# input() lê o que o usuário digitar
nome = input ("Qual é o seu nome?")
print(f"Olá, {nome}! Bem-vindo ao Python.")

# ATENÇÃO: input() SEMPRE retorna texto (str)!
idade = int(input("Sua idade: "))

print(10 + 3) #13 - soma
print(10 - 3) # 7 - subtração
print(10 * 3) #30 - multiplicação
print(10 / 3) #3.33 - divisão
print(10 // 3)#3 - divisão inteira
print(10 % 3) #1 - resto da divisão
print(10 ** 2) #100 - potência (10²)

# Exemple prático: calculadora de IMC
peso   = float(input("Peso (kg): "))
altura = float(input("Altura (m): "))
imc    = peso / altura ** 2
print(f"Seu IMC é: {imc:.1f}")

#Python · O erro mais comum no início

# input() retorna SEMPRE string — cuidado!
idade = input("Sua idade: ")
print(idade + 1)   #  TypeError!

# Solução: converter para int
idade = int(input("Sua idade: "))
print(idade + 1)   #  funciona!
```
