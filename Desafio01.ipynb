```python
nome = input(f"Qual o seu nome?")

prod1 = input(f"\nQual o produto?")
q1 = int (input(f"Qual a quantidade?"))
p1 = input(f"Qual o valor dos(as){prod1}?R$")
p1 = float(p1.replace(',', '.'))
sub1 = (q1 * p1)
print((f"\nsoma total dos(as) {prod1} = R${sub1:.2f}").replace('.',','))

prod2 = input(f"\nQual o produto?")
q2 = int (input(f"Quantos a quantidade?"))
p2 = input(f"Qual o valor dos(as){prod2}?R$")
p2 = float(p2.replace(',', '.'))
sub2 = (q2 * p2)
print((f"\nsoma total dos(as){prod2} = R${sub2:.2f}").replace('.',','))

prod3 = input(f"\nQual o produto?")
q3 = int (input(f"Qual a quantidade?"))
p3 = input(f"Qual o valor dos(as){prod3}?R$")
p3 = float(p3.replace(',', '.'))
sub3 = (q3 * p3)
print((f"\nsoma total do {prod3} = R${sub3:.2f}").replace('.',','))

somat = sub1 + sub2 + sub3

print(f"{somat:.2f}")

desconto = somat * 0.10
p_final = somat - desconto

print(f"n\CUPOM FISCAL")
print(f"{prod1}:unidades:{q1}=R${sub1}")
print(f"{prod2}:unidades:{q2}=R${sub2}")
print(f"{prod3}:unidades:{q3}=R${sub3}")
print(f"Total com desconto de 10%:{p_final:.2f}")
```

```python
lista = [42, 11, 22, 14, 15]

# Imprime apenas valores pares
for valor in lista:
  if valor % 2 == 0:
    print(valor)


# Imprime apenas valores impares
# continue ignora codigo abaixo naquele loop
for valor in lista:
  if valor % 2 == 0:
    continue
  print(valor)
```

```python
print("Calcular média")
lista = [20, 5, 10, 15, 28, 30]
soma = 0
contagem = 0
for elemento in lista:
  contagem = contagem + 1
for valor in lista:
  soma = soma + valor
print(f"A lista contém {contagem} e soma de todos é {soma} ")
media = soma / contagem
print(f"a média da lista é {media}")

```
